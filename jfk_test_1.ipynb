<h2>Задача 1. Найти и вывести долю повторных покупателей от общего числа покупателей в категории. Повторные покупатели -- покупатели, совершившие две и более покупки в текущем периоде времени. </h2>

<p><strong><font color="blue">Определение "текущего периода времени" из задачи вызывает вопрос, но выдвину гипотезу, что речь идет о покупках в один день и далее буду исходить из нее.</font></strong>.</p>

```python
count_purchases = df1.groupby('hhkey')['occaskey'].nunique()
total_customers_one_purchase = count_purchases[count_purchases == 1].count()
print("Всего покупателей, сделавших только один акт покупки:", total_customers_one_purchase)
```

```python
# Группируем данные по hhkey и movedate, подсчитываем количество покупок (occaskey) для каждого покупателя в каждую дату
grouped = df1.groupby(['hhkey', 'movedate'])['occaskey'].count().reset_index()

# Фильтруем данные, оставляем только покупателей с двумя и более покупками в один день
filtered = grouped[grouped['occaskey'] >= 2]

# Считаем количество уникальных покупателей и количество удовлетворяющих условию
total_customers = df1['hhkey'].nunique()
qualified_customers = filtered['hhkey'].nunique()
share = qualified_customers / total_customers
print("Доля покупателей с двумя и более покупками в одну дату:", share)
```

<h2>Задача 2. Разделить повторных покупателей на три группы в зависимости от совокупных затрат в категории: heavy / medium / light buyers. В качестве критерия использовать тертили по тратам в категории за весь период среди всех повторных покупателей. Найти и вывести долю каждой группы покупателей от общего числа повторных покупателей. </h2>

<p><strong><font color="blue">решим эту задачу поэтапно:

Расчет совокупных затрат: Сначала найдем совокупные затраты для каждого повторного покупателя. Для этого умножим полную стоимость товара на количество товаров в покупке.
Определение тертилей: Разделим повторных покупателей на три группы (heavy, medium и light buyers) с использованием тертилей (25%, 50% и 75%) по совокупным затратам.
Подсчет доли групп: Найдем долю каждой группы покупателей от общего числа повторных покупателей.</font></strong>.</p>

```python
print(result_df.info())
```

```python
result_df.loc[:, 'value'] = result_df['value'].astype(float)
result_df.loc[:, 'number'] = result_df['number'].astype(int)
```

```python
# Рассчитываем общие затраты каждого покупателя
result_df.loc[:, 'total_spent'] = result_df['value'] * result_df['number']

# Разделяем повторных покупателей на три группы - heavy / medium / light buyers
# Находим тертили по затратам
tertiles = result_df['total_spent'].quantile([0.33, 0.66])

# Функция для классификации покупателей
def buyer_category(total_spent):
    if total_spent <= tertiles[0.33]:
        return 'light'
    elif total_spent <= tertiles[0.66]:
        return 'medium'
    else:
        return 'heavy'

result_df.loc[:, 'buyer_category'] = result_df['total_spent'].apply(buyer_category)

# Вычисляем долю каждой группы покупателей от общего числа повторных покупателей
buyer_distribution = result_df['buyer_category'].value_counts(normalize=True)

print(buyer_distribution)
```

<p><strong><font color="blue">тот же результат в виде процентов</font></strong>.</p>

```python
buyer_distribution_percent = buyer_distribution * 100

print(buyer_distribution_percent)
```

<p><strong><font color="blue">Для решения этой задачи сделаем следующее:
1. Объединим `result_df` с `data` по колонке `hhkey`, чтобы получить информацию о федеральных окргуах покупателей.
2. Для каждой группы и для каждого региона рассчитаем средний размер корзины (сумма value / сумма number).</font></strong>.</p>

```python
merged_df = pd.merge(result_df, data, on='hhkey', how='inner')
def buyer_category(total_spent):
    if total_spent <= tertiles[0.33]:
        return 'light'
    elif total_spent <= tertiles[0.66]:
        return 'medium'
    else:
        return 'heavy'

result_df.loc[:, 'buyer_category'] = result_df['total_spent'].apply(buyer_category)
result = merged_df.groupby(['federatio', 'buyer_category'])[['value', 'number']].sum()
result['avg_basket_size'] = result['value'] / result['number']

# Вывод результата
print(result)

```

<h2>Задача 4. Построить помесячный график трат в категории в целом и по каналам</h2> 

```python
import matplotlib.pyplot as plt
```

```python
# Выделим месяц из колонки `movedate` и создадим колонку `month`:
df1['movedate'] = pd.to_datetime(df1['movedate'])
df1['month'] = df1['movedate'].dt.to_period('M')
df1['month'] = df1['month'].astype(str)
df1['value'] = pd.to_numeric(df1['value'], errors='coerce')

#Построим помесячный график общих трат:

monthly_total = df1.groupby('month')['value'].sum()
plt.figure(figsize=(12, 6))
plt.plot(monthly_total, marker='o')
plt.title('Total Spending per Month')
plt.xlabel('Month')
plt.ylabel('Total Spending')
plt.grid(True)
plt.show()
#Построим помесячный график трат по каналам:

monthly_channel_spending = df1.groupby(['month', 'channel'])['value'].sum().unstack()
print(monthly_channel_spending)
plt.figure(figsize=(12, 6))
monthly_channel_spending.plot(kind='bar', stacked=True)
plt.title('Ежемесячные расходы по каналам')
plt.xlabel('Месяц')
plt.show()
```
